```python
import pandas
from matplotlib import pyplot
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import numpy
```

```python
f=open('amk.txt','r')
f_x=[]
f_y=[]
ai_x=[]
ai_y=[]
temp_x=[]
temp_y=[]
for line in f:
   line=line.split(',') 
   line=list(map(lambda x:int(x),line))
   f_x.append(line[0])
   f_y.append(line[1])
   ai_x.append(line[2])
   ai_y.append(line[3])
   temp_x.append(line[4])
   temp_y.append(line[5]) 
```

```python
d=({'food_x':f_x,'food_y':f_y,'player_x':ai_x,'player_y':ai_y,'temp_x':temp_x,'temp_y':temp_y})
```

```python
df=pandas.DataFrame(data=d)
df
```

```python
d=[]
for i in range(len(temp_x)):
    if temp_x[i]==0 and temp_y[i]==20:
        d.append('down')
    if temp_x[i]==0 and temp_y[i]==-20:
        d.append('up')
    if temp_y[i]==0 and temp_x[i]==20:
        d.append('right')
    if temp_y[i]==0 and temp_x[i]==-20:
        d.append('left')
df=df.assign(direction=d)
df['d']=LabelEncoder().fit_transform(df['direction'])
df
```

```python
fig=pyplot.figure(figsize=(16,9))
axis=fig.add_axes([1,1,1,1])
axis.scatter(f_x,f_y,color='r')
axis.plot(ai_x,ai_y)
axis.legend(['player_position','food_position'])
```

```python
x=numpy.array(df.iloc[:,:4]).reshape(2510,4)
y=numpy.array(df.iloc[:,7])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
tree=DecisionTreeClassifier(max_depth=12)
tree.fit(x_train,y_train)
tree.score(x_train,y_train)
```

```python
import pygame
import random
import math

white=(255,255,255)
black=(0,0,0)
red=(255,0,0)
green=(0,255,0)
blue=(0,0,255)

pygame.init()

display_width=800
display_height=600
game_display=pygame.display.set_mode((display_width,display_height))
pygame.display.set_caption('AMK OS')
clock=pygame.time.Clock()
FPS=30

class snake:
      def draw(self,color):
           pygame.draw.circle(game_display,color,[self.x,self.y],10)

      def update(self,snake_obj):
           self.x=snake_obj.x 
           self.y=snake_obj.y     
    
                
class snake_ai(snake):
      def __init__(self):
          self.x=random.randrange(display_width)
          self.y=random.randrange(display_height)
          self.temp_x=20
          self.temp_y=0

      def update(self,food_obj):
          y_pred=[0]
          if self.x-food_obj.x<=17 or self.y-food_obj.y<=17:
            y_pred=tree.predict(numpy.array(numpy.array([food_obj.x,food_obj.y,self.x,self.y]).reshape(-1,4)))
          
          if y_pred[0]==0:#down
             self.temp_x=0
             self.temp_y=20
          elif y_pred[0]==1:#left
             self.temp_x=-20
             self.temp_y=0
          elif y_pred[0]==2:#right
             self.temp_x=20
             self.temp_y=0
          elif y_pred[0]==3:#up
            self.temp_x=0
            self.temp_y=-20
          self.x+=self.temp_x
          self.y+=self.temp_y
          
          if self.x<0:
                self.x=display_width
          elif self.x>display_width:
                self.x=0
          elif self.y<0:
                self.y=display_height
          elif self.y>display_width:
                self.y=0   
  
          #print(y_pred)  

          

class food:
      def __init__(self):
            self.r=10
            self.x=random.randrange(5,display_width-5)
            self.y=random.randrange(5,display_height+5)

      def draw(self):
            pygame.draw.circle(game_display,red,[self.x,self.y],5)

      def collision_detect(self,snake_obj):
           if math.sqrt((self.x-snake_obj.x)**2+(self.y-snake_obj.y)**2)<=15:
              self.x=random.randrange(display_width)
              self.y=random.randrange(display_height)
              return True
           return False 
                
            

#player=[snake_head()]
ai=[snake_ai()]
f=food()


game_exit='false'
  
    
while game_exit!=True:

    clock.tick(FPS)
    game_display.fill(white)

    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            game_exit=True

    ai[0].draw(blue)
    ai[0].update(f)#movement of ai 
    f.draw()

    if f.collision_detect(ai[0]):#collision detection with ai
       ai.append(snake())

    i=len(ai)-1
    while i>=1:#draw ai    
        ai[i].update(ai[i-1])
        ai[i].draw(blue)
        i-=1    
    pygame.display.flip()

pygame.quit()

    
```
